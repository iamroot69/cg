<h2>Bresenham LDA</h2>

```python
x1,y1,x2,y2=eval(input("Enter points"))
del_x=x2-x1
del_y=y2-y1
l_x=[]
l_y=[]
m = (y2 - y1) / (x2 - x1)
P = 2 * del_y - del_x
x = x1
y = y1
while y2 != y and x2 != x:
    # print(x,y)
    l_x.append(x)
    l_y.append(y)
    if m<1:
        x=x+1
        if P<0:
            P = P + 2 * del_y
        else:
            y=y+1
            P = P + 2 * (del_y - del_x)
    else:
        y=y+1
        if P<0:
            P = P + 2 * del_x
        else:
            x=x+1
            P = P + 2 * (del_x - del_y)
# print(x,y)
l_x.append(x)
l_y.append(y)
if x!=x2 or y!=y2:
    l_x.append(x2)
    l_y.append(y2)
```

```python
import matplotlib.pyplot as plt
```

```python
plt.xlim(max(x2,y2))
plt.ylim(max(x2,y2))
plt.scatter(l_x,l_y)
```

<h2>DDA LDA</h2>

```python
l_x=[]
l_y=[]
x=x1
y=y1
while y2 != round(y) or x2 != round(x):
    print(x,y)
    if abs(del_x)>=abs(del_y):
        x=x+1
        y=y+m
    else:
        x=x+1/m
        y=y+1
    l_x.append(round(x))
    l_y.append(round(y))

if round(x)!=x2 or round(y)!=y2:
    l_x.append(x2)
    l_y.append(y2)
```

```python
plt.xlim(max(x2,y2))
plt.ylim(max(x2,y2))
plt.scatter(l_x,l_y)
```
